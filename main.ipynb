```python
import numpy as np

def bisection_method(f, a, b, tol):
    if f(a) * f(b) >= 0:
        print("El teorema de bisecci√≥n no es aplicable.")
        return None
    else:
        while (b - a) / 2 > tol:
            c = (a + b) / 2
            if f(c) == 0:
                return c
            elif f(a) * f(c) < 0:
                b = c
            else:
                a = c
        return (a + b) / 2

# (a) f(x) = x^2 - 2 en el intervalo [0, 4]
def f_a(x):
    return x**2 - 2

sol_a = bisection_method(f_a, 0, 4, 0.005)
print(f"Soluci√≥n para (a): x ‚âà {sol_a}")

# (b) f(x) = x^5 - 6x^3 + 2x - 2 en el intervalo [-1, 1]
def f_b(x):
    return x**5 - 6*x**3 + 2*x - 2

sol_b = bisection_method(f_b, -1, 1, 0.005)
print(f"Soluci√≥n para (b): x ‚âà {sol_b}")

# (c) f(x) = cos(x) - x en el intervalo [0, œÄ]
def f_c(x):
    return np.cos(x) - x

sol_c = bisection_method(f_c, 0, np.pi, 0.005)
print(f"Soluci√≥n para (c): x ‚âà {sol_c}")

# (d) f(x) = 2sin(x) - sqrt(x + 1) en el intervalo [0, 2]
def f_d(x):
    return 2*np.sin(x) - np.sqrt(x + 1)

sol_d = bisection_method(f_d, 0, 2, 0.005)
print(f"Soluci√≥n para (d): x ‚âà {sol_d}")
```

#### Definir la funci√≥n $ h(x) $ como la diferencia entre $ f(x) $ y $ g(x) $:

$
h(x) = f(x) - g(x)
$

Dado que $ f(x) $ y $ g(x) $ son continuas en el intervalo $ [0, 1] $, entonces $ h(x) $ tambi√©n es continua en $ [0, 1] $ porque es la diferencia de dos funciones continuas.

#### Evaluar $ h(x) $ en los extremos del intervalo

Sabemos que $ f(0) < g(0) $ y $ f(1) > g(1) $, por lo que:

$
h(0) = f(0) - g(0) < 0
$
$
h(1) = f(1) - g(1) > 0
$

####  Aplicar el Teorema del Valor Intermedio (TVI)

El Teorema del Valor Intermedio dice que si una funci√≥n continua cambia de signo en un intervalo cerrado, entonces existe un punto en el intervalo donde la funci√≥n es cero.

En este caso, $ h(x) $ es continua en $ [0, 1] $, y tenemos que:

$
$h(0) < 0 \quad \text{y} \quad h(1) > 0$
$

Por lo tanto, por el Teorema del Valor Intermedio, debe existir un punto $ c \in (0, 1) $ tal que:

$
h(c) = 0
$

#### Concluir que $ f(c) = g(c) $

Si $ h(c) = 0 $, entonces:

$
f(c) - g(c) = 0 \quad \Rightarrow \quad f(c) = g(c)
$

Conclusi√≥n

Existe un punto $ c \in (0, 1) $ tal que $ f(c) = g(c) $.
