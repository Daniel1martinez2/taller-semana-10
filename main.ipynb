$
f(t) = \frac{t^2 - 7t + 12}{t - 4}
$

Primero, factorizamos el numerador \( t^2 - 7t + 12 \):

$
t^2 - 7t + 12 = (t - 3)(t - 4)
$

Por lo tanto, la función original se puede reescribir como:

$
f(t) = \frac{(t - 3)(t - 4)}{t - 4}
$

Siempre que \( t- 4 \), podemos cancelar el término \( t - 4 \) en el numerador y el denominador:

$
f(t) = t - 3
$

Sin embargo, en \( t = 4 \), la función original no está definida, ya que estamos dividiendo por cero. Esto provoca una **discontinuidad en \( t = 4 \)**.

La discontinuidad en \( t = 4 \) es **removible** porque la función simplificada \( f(t) = t - 3 \) tiene un valor bien definido cuando \( t = 4 \). Para hacer que la función sea continua en \( t = 4 \), podemos definir una nueva función \( g(t) \) de la siguiente forma:

$
g(t) =
\begin{cases}
t - 3 & \text{si } t \neq 4, \\
1 & \text{si } t = 4.
\end{cases}
$

De esta manera, \( g(t) \) es continua en todo punto, incluyendo \( t = 4 \).

La discontinuidad en \( t = 4 \) es removible. La nueva función \( g(t) \) es continua en \( t = 4 \) porque el valor de la función en ese punto coincide con el límite de la función cuando \( t \) tiende a 4:

$
\lim_{t \to 4} f(t) = 1
$

```python
import numpy as np

def bisection_method(f, a, b, tol):
    if f(a) * f(b) >= 0:
        print("El teorema de bisección no es aplicable.")
        return None
    else:
        while (b - a) / 2 > tol:
            c = (a + b) / 2
            if f(c) == 0:
                return c
            elif f(a) * f(c) < 0:
                b = c
            else:
                a = c
        return (a + b) / 2

# (a) f(x) = x^2 - 2 en el intervalo [0, 4]
def f_a(x):
    return x**2 - 2

sol_a = bisection_method(f_a, 0, 4, 0.005)
print(f"Solución para (a): x ≈ {sol_a}")

# (b) f(x) = x^5 - 6x^3 + 2x - 2 en el intervalo [-1, 1]
def f_b(x):
    return x**5 - 6*x**3 + 2*x - 2

sol_b = bisection_method(f_b, -1, 1, 0.005)
print(f"Solución para (b): x ≈ {sol_b}")

# (c) f(x) = cos(x) - x en el intervalo [0, π]
def f_c(x):
    return np.cos(x) - x

sol_c = bisection_method(f_c, 0, np.pi, 0.005)
print(f"Solución para (c): x ≈ {sol_c}")

# (d) f(x) = 2sin(x) - sqrt(x + 1) en el intervalo [0, 2]
def f_d(x):
    return 2*np.sin(x) - np.sqrt(x + 1)

sol_d = bisection_method(f_d, 0, 2, 0.005)
print(f"Solución para (d): x ≈ {sol_d}")
```

- Solución para (b): x ≈ -0.90234375

- Solución para (d): x ≈ 0.71484375

```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.1, 20, 400)

D = (50000 * p + 10000) / (p**2)
S = (1/3) * p**2 + 2 * p

# Graficar
plt.plot(p, D, label="Demanda D(p)")
plt.plot(p, S, label="Oferta S(p)")
plt.xlabel("Precio (p)")
plt.ylabel("Unidades")
plt.legend()
plt.title("Demanda y Oferta en el mismo gráfico")
plt.grid(True)
plt.show()
```

```python
##Punto b
def D(p):
    return (50000 * p + 10000) / (p ** 2)

def S(p):
    return (1/3) * p**2 + 2 * p

def f(p):
    return D(p) - S(p)

def bisection(a, b, tol=1e-5):
    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        if f(midpoint) == 0:
            return midpoint
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
    return (a + b) / 2.0

equilibrium_price = bisection(1, 20)
print("Precio de equilibrio estimado:", equilibrium_price)

##Punto c
equilibrium_units = S(equilibrium_price)
print("Unidades producidas al precio de equilibrio:", equilibrium_units)
```

La función de ganancia es:

$P(p) = I(p) - C(p) = (750p - 50p^2) - (3750 - 250p) = -50p^2 + 1000p - 3750$

Para maximizar la ganancia, derivamos $P(p)$ respecto a  $p$, usando la notación $P'(p)$:

$P'(p) = -100p + 1000$

Igualamos la derivada a cero para encontrar el valor que maximiza la ganancia:


$-100p + 1000 = 0$

$p = 10$

Por lo tanto, el precio que maximiza la ganancia es $p = 10 $.

#### Definir la función $ h(x) $ como la diferencia entre $ f(x) $ y $ g(x) $:

$
h(x) = f(x) - g(x)
$

Dado que $ f(x) $ y $ g(x) $ son continuas en el intervalo $ [0, 1] $, entonces $ h(x) $ también es continua en $ [0, 1] $ porque es la diferencia de dos funciones continuas.

#### Evaluar $ h(x) $ en los extremos del intervalo

Sabemos que $ f(0) < g(0) $ y $ f(1) > g(1) $, por lo que:

$
h(0) = f(0) - g(0) < 0
$
$
h(1) = f(1) - g(1) > 0
$

####  Aplicar el Teorema del Valor Intermedio (TVI)

El Teorema del Valor Intermedio dice que si una función continua cambia de signo en un intervalo cerrado, entonces existe un punto en el intervalo donde la función es cero.

En este caso, $ h(x) $ es continua en $ [0, 1] $, y tenemos que:

$
$h(0) < 0 \quad \text{y} \quad h(1) > 0$
$

Por lo tanto, por el Teorema del Valor Intermedio, debe existir un punto $ c \in (0, 1) $ tal que:

$
h(c) = 0
$

#### Concluir que $ f(c) = g(c) $

Si $ h(c) = 0 $, entonces:

$
f(c) - g(c) = 0 \quad \Rightarrow \quad f(c) = g(c)
$

Conclusión

Existe un punto $ c \in (0, 1) $ tal que $ f(c) = g(c) $.
